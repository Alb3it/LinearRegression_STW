```python
import pandas as pd
file = pd.read_csv("무선이어폰 분실수.txt", delimiter=' ')
file
```

```python
import matplotlib.pyplot as plt

plt.rc('xtick', labelsize=5)  # x축 눈금 폰트 크기


plt.title('Dispersion of the number of lost wireless earphones by week')
plt.xlabel('Week(start-date_end-date)')
plt.ylabel('number of wireless earphones lost')
plt.scatter(file['week'], file['number_of_earphones'])
```

```python
during_retrieve = file.loc[8:16]
during_retrieve
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
lf = LinearRegression()
x = [[0], [1], [2], [3], [4], [5], [6], [7], [8]]
lf.fit(x, during_retrieve['number_of_earphones'])
```

```python
lf.coef_
```

```python
plt.plot(x, during_retrieve['number_of_earphones'], 'o')
plt.plot(x, lf.predict(x))
plt.xlabel('index of weeks')
plt.ylabel('number of earphones')
plt.title('Linear Regression in retrieve period')
```

```python
lf.score(x, during_retrieve['number_of_earphones'])
```
